```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path= 'Data Analysis - Caselet#2.xlsx'
# Assuming your data is in an Excel file named "your_data_file.xlsx"
df = pd.read_excel(file_path, sheet_name='DifferentFormat')

# Plotting using Seaborn
plt.figure(figsize=(14, 8))
sns.boxplot(x='Team Member', y='TPT', data=df)
plt.title('Box Plot of TPT by Team Member')
plt.xlabel('Team Member')
plt.ylabel('TPT')
plt.xticks(rotation=45, ha="right")
plt.show()
```

- Check for variations in TPT among different Team Members.
- Identify any outliers or extreme values.
